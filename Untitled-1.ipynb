```python
# Define a function to determine if a food is healthy
def is_healthy(row):
    # Define vitamin thresholds (adjust as needed)
    min_vitamin_a = 500  # mcg
    min_vitamin_c = 30   # mg
    min_vitamin_b1 = 0.2 # mg
    min_vitamin_b2 = 0.2 # mg
    min_vitamin_b3 = 2   # mg
    min_folate = 50      # mcg
    
    # Check if the food meets at least one of these conditions
    if (row['Vitamin A (mcg)'] >= min_vitamin_a or
        row['Vitamin C (mg)'] >= min_vitamin_c or
        row['Vitamin B1 (mg)'] >= min_vitamin_b1 or
        row['Vitamin B2 (mg)'] >= min_vitamin_b2 or
        row['Vitamin B3 (mg)'] >= min_vitamin_b3 or
        row['Folate (mcg)'] >= min_folate):
        return 'Yes'
    else:
        return 'No'

# Apply the function to create the 'healthy_food' column
df['healthy_food'] = df.apply(is_healthy, axis=1)

df.head()
```

```python
df.fillna(0, inplace=True)  # Replace NaNs with 0

```
